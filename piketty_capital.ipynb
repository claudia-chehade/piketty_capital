```python
quandl.ApiConfig.api_key = "XXX"
```

```python
def plot_time_series(df_data, title):
    df_data.plot(title=title, 
                         figsize=(8, 8), 
                         legend=True);
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
    plt.show();
```

```python
def load_and_display(table_name, title, start_date=None, end_date=None):
    if start_date and end_date:
        df_data = quandl.get(table_name, start_date=start_date, end_date=end_date)
    else:
        df_data = quandl.get(table_name)
    plot_time_series(df_data, title)
    return df_data
```
